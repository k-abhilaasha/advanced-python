```python
import random
p1=input("enter player1: ")
p2=input("enter player2: ")
s1=10
s2=10
d1=random.randint(1,10)
d2=random.randint(1,10)
print("Player1 turn")
while(True):
  g1=int(input("Enter ur guess:"))
  s1=s1-1
  if g1==d1:
    break
print("Player2 turn")
while(True):
  g2=int(input("Enter ur guess:"))
  s2=s2-1
  if g2==d2:
    break
print("Summary")
print("{}:{}".format(p1,s1))
print("{}:{}".format(p2,s2))
print("-------")
if s1>s2:
  print("Winner is ",p1)
elif s2>s1:
  print("Winner is",p2)
else:
  print("Draw")
```

```python
#A poultry farm has group of hens and cows..need to find num of cows and hens separately if total legs and total heads together given:
legs=int(input("Enter No of legs:"))
heads=int(input("Enter No of heads:"))
temp=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4+hens*2
  if(cal_legs==legs):
    temp=True
    break
if temp:
  print("Cows:",cows,"\nHens:",hens)
else:
  print("No Solution")
```
